```python
## Creating an External Python file for SI Calculation.
with open('si.py','w') as f:
    f.write("""
def  calc_si(principal,rate,time):
    si = (principal * rate * time) / 100
    return si
def result(principal,rate,time):
    intr = calc_si(principal,rate,time)
    total_amt = principal + intr

    print("Principal Amount :",principal)
    print("Rate of Interest : ",rate,"%")
    print("Time : ",time,"Years")
    print("Simple Interest :",intr)
    print("Total Amount : ",total_amt)
""")
```
